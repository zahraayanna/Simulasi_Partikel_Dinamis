```python
import numpy as np

n_particles = 50      #jumlah partikel
box_width = 10        #ukuran kotak
n_steps = 10000       #jumlah iterasi
dt = 0.001            #interval waktu
global trajectory
```

```python
def get_initial_coordinates():
  x_coord = [np.random.random()*box_width for i in range(n_particles)]
  y_coord = [np.random.random()*box_width for i in range(n_particles)]

  return x_coord, y_coord
```

```python
def get_initial_velocities():
    x_vel = [(np.random.random() - 0.5) * 2 for i in range(n_particles)]
    y_vel = [(np.random.random() - 0.5) * 2 for i in range(n_particles)]
    return x_vel, y_vel
```

```python
def take_steps(x_coord, y_coord,x_vel, y_vel):
  for i in range(n_particles):
    x_coord[i] += x_vel[i]*dt
    y_coord[i] += y_vel[i]*dt

    if x_coord[i] < 0 or x_coord[i] > box_width:
      x_vel[i] = -x_vel[i]
    if y_coord[i] < 0 or y_coord[i] > box_width:
      y_vel[i] = -y_vel[i]

  return x_coord, y_coord, x_vel, y_vel
```

```python
def add_frame(xs,ys,i):
  global trajectory
  if i ==0:
    trajectory =''
  trajectory += str(n_particles)+ '\ntitle\n'
  for x, y in zip(xs,ys):
    trajectory += ' '.join(['Ar',str(x),str(y),'0.0\n'])
```

```python
x_coord, y_coord = get_initial_coordinates()

x_vel, y_vel = get_initial_velocities()

for i in range(n_steps):
  x_coord, y_coord, x_vel, y_vel = take_steps(x_coord, y_coord, x_vel, y_vel)

  if i%10 == 0:
    add_frame(x_coord, y_coord,i)
```
